**Find the python program to find the odd numbers in the list of numbers entered by user.**

```python
numbers = input("Enter numbers separated by spaces: ").split()
odd_numbers = [int(num) for num in numbers if int(num) % 2 == 1]

print("Odd numbers:", odd_numbers)
```

**Write a python & create a class named Employee with attributes name, role, and salary. Write a method to calculate the annual salary of the employee.**

```python
class Employee:
    def __init__(self, name, role, salary):
        self.name = name
        self.role = role
        self.salary = salary

    def annual_salary(self):
        return self.salary * 12

employee = Employee("Khuteeja B", "Quality Engineer", 50000)
print(f"Annual Salary of {employee.name}: {employee.annual_salary()}")
```

**Write a python program to convert colored image to grayscale and rotate it in clockwise 90 degrees using opencv**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

image = cv2.imread('img.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

height, width = gray_image.shape[:2]
center = (width // 2, height // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, -90, 1.0)  # -90 for clockwise rotation

rotated_image = cv2.warpAffine(gray_image, rotation_matrix, (width, height))

cv2_imshow( image)
cv2_imshow(gray_image)
cv2_imshow(rotated_image)
```

**Given a dataset containing exam scores of students, write a program using NumPy to perform the following statistical analysis: a) Calculate the mean and median of the exam scores. b) Compute the standard deviation and variance of the scores.**

```python
import numpy as np

scores = np.array([85, 92, 78, 95, 88, 70, 90, 82])

# a) Calculate mean and median
mean_score = np.mean(scores)
median_score = np.median(scores)

print(f"Mean score: {mean_score}")
print(f"Median score: {median_score}")

# b) Compute standard deviation and variance
std_deviation = np.std(scores)
variance = np.var(scores)

print(f"Standard Deviation: {std_deviation}")
print(f"Variance: {variance}")
```

**Write a Python Program to Multiply Two Matrices and check resultant matrix order.**

```python
import numpy as np

# Define two matrices
A = np.array([
    [1, 2],
    [3, 4],
    [5, 6]
])

B = np.array([
    [7, 8, 9],
    [10, 11, 12]
])

# Multiply the matrices
result = np.dot(A, B)

# Get the order of the resultant matrix
order = result.shape

print("Resultant Matrix:")
print(result)
print(f"Order of Resultant Matrix: {order}")
```
